```python
X = df.drop("selling_price", axis=1)
y = df["selling_price"]
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=15)
```

```python
cat_cols = df.select_dtypes("object").columns.to_list()
```

```python
cat_cols
```

```python
unique_values = df[cat_cols].nunique()
```

```python
unique_values
```

```python
# seller_type, fuel_type, transmission_type -> onehotencoding
# car_name, brand, model -> frequency encoding
```

```python
onehot_columns = ["seller_type", "fuel_type", "transmission_type"]
freq_columns = ["car_name", "brand", "model"]
```

```python
for col in freq_columns:
    freq=X_train[col].value_counts() / len(X_train)

    X_train[col + '_freq'] = X_train[col].map(freq)
    X_test[col + '_freq'] = X_test[col].map(freq)
    mean_freq = freq.mean()
    X_test[col + '_freq'] = X_test[col + '_freq'].fillna(mean_freq)
```

```python
X_train.head()
```

```python
X_test.head()
```

```python
X_train = X_train.drop(["car_name", "brand", "model"], axis=1)
X_test = X_test.drop(["car_name", "brand", "model"], axis=1)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
transformer = ColumnTransformer(
    transformers = [
        ("onehot", OneHotEncoder(drop="first", handle_unknown="ignore"),onehot_columns)
    ], remainder = "passthrough"
)
```

```python
X_train = transformer.fit_transform(X_train)
X_test = transformer.transform(X_test)
```

```python
encoded_cols = transformer.get_feature_names_out()
```

```python
encoded_cols
```

```python
X_train = pd.DataFrame(X_train, columns=encoded_cols)
X_test = pd.DataFrame(X_test, columns=encoded_cols)
```

```python
X_train.head()
```

```python
X_train.info()
```

```python
from sklearn.ensemble import AdaBoostRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from sklearn.model_selection import RandomizedSearchCV
```

```python
model = AdaBoostRegressor()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)


```

```python
print("r2_score:", r2_score(y_pred, y_test))
print("mean squared error:", mean_squared_error(y_pred, y_test))
print("mean absolute error:", mean_absolute_error(y_pred, y_test))
```

```python
params = {
    "n_estimators" : [50,80,100,120],
    "learning_rate" : [0.001, 0.01, 0.1, 1.0, 2.0],
    "loss" : ["linear", "square", "exponential"]
}
```

```python
rcv = RandomizedSearchCV(AdaBoostRegressor(), param_distributions=params, scoring="r2", cv=5)
```

```python
rcv.fit(X_train, y_train)
```

```python
rcv.best_params_
```

```python
y_pred = rcv.predict(X_test)
```

```python
print("r2_score:", r2_score(y_pred, y_test))
print("mean squared error:", mean_squared_error(y_pred, y_test))
print("mean absolute error:", mean_absolute_error(y_pred, y_test))
```

```python
params = {
    "estimator__max_depth" : [3,4,5],
    "n_estimators" : [50,80,100,120],
    "learning_rate" : [0.001, 0.01, 0.1, 1.0, 2.0],
    "loss" : ["linear", "square", "exponential"]
}
```

```python
rcv = RandomizedSearchCV(AdaBoostRegressor(DecisionTreeRegressor()), param_distributions=params, scoring="r2", cv=5)
```

```python
rcv.fit(X_train, y_train)
```

```python
rcv.best_params_
```

```python
y_pred = rcv.predict(X_test)
print("r2_score:", r2_score(y_pred, y_test))
print("mean squared error:", mean_squared_error(y_pred, y_test))
print("mean absolute error:", mean_absolute_error(y_pred, y_test))
```
